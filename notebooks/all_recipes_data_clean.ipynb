```python
import pandas as pd
import numpy as np 
# data = pd.read_csv("data/test_recipes.csv")
original_data = pd.read_csv("../data/recipes.csv")
data = original_data[['recipe_name', 'prep_time', 'cook_time', 'total_time', 'servings', 'yield', 'ingredients', 'directions', 'url']].copy()
data.rename(columns={"recipe_name" : "Name", "prep_time" : "Prep Time", "cook_time" : "Cook Time", "total_time" : "Total Time", "servings" : "Servings", "yield" : "Yield", "ingredients" : "Ingredients", "directions" : "Directions"}, inplace=True)
data.head(10)
```

```python
data.info()
```

```python
# clean dataset

data = data.drop(data.columns[0], axis=1)
```

```python
def convert_time_to_mins(time):
    if pd.isna(time):
        return np.nan
    
    time = time.lower().strip()
    split = time.split()

    # 3 possible formats:
    # x hrs y mins
    if len(split) == 4:
        return int(split[0]) * 60 + int(split[2])

    # x hrs
    if split[1] == "hrs":
        return int(split[0]) * 60
    
    # y mins
    if split[1] == "mins":
        return int(split[0])    
    
    return np.nan # shouldn't get here

# testing
print(convert_time_to_mins("1 hrs"))          # 120
print(convert_time_to_mins("40 mins"))        # 45
print(convert_time_to_mins("1 hrs 5 mins"))  # 90
```

```python
data["Prep Time"] = data["Prep Time"].apply(convert_time_to_mins).astype("Int64")
data["Cook Time"] = data["Cook Time"].apply(convert_time_to_mins).astype("Int64")
data["Total Time"] = data["Total Time"].apply(convert_time_to_mins).astype("Int64")

data.head(10)
data.info()
```

```python
data['Servings'] = data['Servings'].astype('Int64')

print(data['Yield'].unique())
data = data.drop('Yield', axis=1)
data.info()
```
